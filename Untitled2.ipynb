```python
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
df.columns = ('Model', 'Mileage', 'Price_dollar', 'Age(yrs)')
df.columns
```

```python
df
```

# Using Pandas to create dummy variables

```python
dummies = pd.get_dummies(df.Model)
dummies
```

```python
merged = pd.concat([df, dummies], axis = 'columns')
merged
```

```python
df_1 = merged.drop(['Model'], axis = 'columns')
df_1
```

# Dummy Variable Trap

```python
df_1 = df_1.drop(['Audi A5'], axis = 'columns')
df_1
```

```python
X = df_1.drop('Price_dollar', axis = 'columns')
X
```

```python
Y = df_1.Price_dollar
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.predict(X)
```

```python
model.score(X,Y)
```

```python
model.predict([[45000, 4, 0, 1]]) #45000 mileage of Mercedez Benz C class that is 4 years old
```

```python
model.predict([[86000, 7, 1, 0]]) #86000 mileage of BMW X5 that is 7 yrs old
```

# Using SKlearn OneHotEncoder

First step is to use label encoder to convert car names into numbers

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.Model = le.fit_transform(dfle.Model)
dfle
```

```python
X = dfle[['Model', 'Mileage', 'Age(yrs)']].values
```

```python
X
```

```python
Y = dfle.Price_dollar.values
Y
```

Now, use oneHotEncoder to create dummy variables for each of the Model


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Model', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
#X.toarray()
```

```python
X = X[:, 1:]
```

```python
X
```

```python
model.fit(X,Y)
```

```python
model.predict([[2, 0, 45000, 4]]) #45000 mileage of Mercedez Benz C class that is 4 years old
```

```python
model.predict([[1, 0, 86000, 7 ]]) #86000 mileage of BMW X5 that is 7 yrs old
```
