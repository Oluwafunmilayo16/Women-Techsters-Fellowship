```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading/reading the dataset
Netflix_data=pd.read_csv('netflix_titles.csv')
Netflix_data
```

```python
#cleaning the dataset_dropping some colums (eg.date_added)
# dropping the few null rows
Netflix_data = Netflix_data.dropna(subset=['date_added', 'rating', 'duration'])
Netflix_data.head()
```

```python
#cleaning the dataset_filling some rows(eg. diector)
Netflix_data['director']=Netflix_data['director'].fillna('Non specified')
Netflix_data
```

```python
#cleaning the dataset_filling null values(eg cast rolls, country rows)
Netflix_data['cast']=Netflix_data['cast'].fillna('Non listed')
Netflix_data['country']=Netflix_data['country'].fillna('Non specified')
Netflix_data
```

```python
#To include only TV show in the dataset
Netflix_TV=Netflix_data[Netflix_data['type']=='TV Show']
Netflix_TV
```

```python
#Top directors for TV show
TV_director=Netflix_TV['director'].str.split(' ',expand=True).stack()
TV_director
```

```python
TV_director=pd.DataFrame(TV_director)
TV_director.head()
```

```python
#Rename the TV_director as Director
TV_director.columns=['Directors']
#Choosing only known TV_directors
TV_director=TV_director[TV_director['Directors']!= 'Non specified']
TV_director=TV_director.groupby(['Directors']).size().reset_index(name='Total TV Show')
top_directors = pd.DataFrame(TV_director.sort_values(by=['Total TV Show'], ascending=False)).head(10)
top_directors
```

```python
#ploting the top directots vs the total TV Show they produced
sns.barplot(y='Directors', x='Total TV Show', data=top_directors, color='red')
plt.title('Top TV Show directors')
```

```python
# Converting the duration to an interger
Netflix_data['duration'] = Netflix_data['duration'].str.strip('min')
Netflix_data['duration'] = Netflix_data['duration'].str.strip('Seasons')
Netflix_data['duration'] = Netflix_data['duration'].str.strip('Seaso')
```

```python
#removing extra white space and converting to integer
Netflix_data['duration']=Netflix_data['duration'].str.strip('').astype('int')
Netflix_data['duration'].value_counts()
```
