```python
import pandas as pd
```

```python
football=pd.read_csv('fifa_ranking_2022-10-06.csv')
football.isnull().sum()
```

```python
football_2=pd.read_csv('matches_1930_2018.csv')
football_2.isnull().sum()
```

```python
football_3=pd.read_csv('matches_2022.csv')
football_3.isnull().sum()
```

```python
football_4=pd.read_csv('world_cup.csv')
football_4.isnull().sum()
```

Merging all our datasets

```python
match=pd.concat([football_2,football_3])
```

```python
match_2=pd.concat([football,football_4])
```

```python
match_final=pd.concat([match,match_2])
match_final
```

```python
match_final['home_penalty'].isnull().sum()
```

checking and filling null values

```python
match_final.isnull().sum()
```

```python
match_final['home_team']=match_final['home_team'].fillna('none_specified')
match_final['home_team']
match_final['away_team']=match_final['away_team'].fillna('none_specified')
match_final['away_team']  
match_final['home_manager']=match_final['home_manager'].fillna('none_specified')
match_final['home_manager']  
match_final['home_captain']=match_final['home_captain'].fillna('none_specified')
match_final['home_captain']  
match_final['away_manager']=match_final['away_manager'].fillna('none_specified')
match_final['away_manager']  

match_final['away_captain']=match_final['away_captain'].fillna('none_specified')
match_final['away_captain']  
match_final['Venue']=match_final['Venue'].fillna('none_specified')
match_final['Venue']  

match_final['Officials']=match_final['Officials'].fillna('none_specified')
match_final['Officials']  

match_final['Date']=match_final['Officials'].fillna('none_specified')
match_final['Date']  

match_final['Referee']=match_final['Referee'].fillna('none_specified')
match_final['Referee']  

match_final['Notes']=match_final['Notes'].fillna('none_specified')
match_final['Notes']  

match_final['Host']=match_final['Host'].fillna('none_specified')
match_final['Host'] 

match_final['Year']=match_final['Year'].fillna('none_specified')
match_final['Year'] 

```

```python
import math
median_home_score=math.floor(match_final.home_score.median())
match_final.home_score=match_final.home_score.fillna(median_home_score)
median_home_xg=math.floor(match_final.home_xg.median())
match_final.home_xg=match_final.home_xg.fillna(median_home_xg)
median_away_score =math.floor(match_final.away_score .median())
match_final.away_score =match_final.away_score.fillna(median_away_score)
median_away_xg=math.floor(match_final.away_xg.median())
match_final.away_xg=match_final.away_xg.fillna(median_away_xg)
median_Attendance =math.floor(match_final.Attendance .median())
match_final.Attendance =match_final.Attendance.fillna(median_Attendance)

```

```python
match_final["home_penalty"].fillna(0,inplace=True)
match_final["away_penalty"].fillna(0,inplace=True)
match_final["away_score"].fillna(0,inplace=True)
match_final["home_goals"].fillna(0,inplace=True)
match_final["away_goals"].fillna(0,inplace=True)
match_final["Score"].fillna(0,inplace=True)
match_final["Round"].fillna(0,inplace=True)
```
