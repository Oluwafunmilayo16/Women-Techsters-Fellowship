```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
HR=pd.read_csv('HR_comma_sep.csv')
HR.head(10)
```

```python
HR.isnull().sum()
```

```python
#getting department
dummies_1=pd.get_dummies(HR.Department)
dummies_1
```

```python
#getting salary
dummies=pd.get_dummies(HR.salary)
dummies
```

```python
#joining department
join_1=pd.concat([HR,dummies_1],axis='columns')
join_1
```

```python
#joining with salary
join_2=pd.concat([join_1,dummies],axis='columns')
join_2
```

```python
HR_1=join_2.drop(['Department'], axis='columns')
HR_1
```

```python
HR_2=HR_1.drop(['salary'], axis='columns')
HR_2
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(HR_2.corr(), annot=True, linewidths=.5, ax=ax)
```

(1)Now do some exploratory data analysis to figure out which variables have direct and clear impact on employee retention (i.e. whether they leave the company or continue to work)

From the heatmap analysis, we obtain a high correlation of 0.85 between low and medium. Hence we drop those columns. This means that those emplyees recieving low salary are leaving the company


```python
#droping those columns
HR_4=HR_2.drop(['low'], axis='columns')
HR_4
```

### 2. Plot bar charts showing impact of employee salaries on retention

```python
#creating a dataframe
df2=pd.DataFrame(HR_4,columns=['time_spend_company','high','medium'])
df2
```

```python
#to determine the correlation between those columns
df2.corr()
```

```python
df2.plot(x='time_spend_company',y=['high','medium'],kind='bar',figsize=(9,8));
plt.xlabel('salary',fontsize=20)
plt.ylabel('count',fontsize=20)
```
